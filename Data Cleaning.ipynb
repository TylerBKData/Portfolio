```python
import pandas as pd 
df_call = pd.read_excel('Customer Call List.xlsx')
df_call

#Need to clean this file
#Look at phone number column, separated by different deliminiters or Paying customer has Yes/No but also Y and N
#Last name has random characters/symbols that shouldn't be there
```

```python
#Remove duplicates
df_call = df_call.drop_duplicates()
df_call #Updated dataframe with a removed duplicate row (has to be exactly the same)
```

```python
df_call = df_call.dropna() #removes all rows with any null values
df_call
```

```python
df_call.loc[df_call['Last_Name'].str.contains('[^(a-zA-Z0-9)]')] #Checks any rows that have characters NOT included in the criteria
```

```python
df_call = df_call.drop(['Not_Useful_Column'], axis=1)
df_call
```

```python
#df_call['Last_Name'] = df_call['Last_Name'].str.lstrip('/') #Removes / from this column on the left
#df_call['Last_Name'] = df_call['Last_Name'].str.lstrip('.') #Removes ... from this column on the left
#df_call['Last_Name'] = df_call['Last_Name'].str.rstrip('_') #Removes _ from this column on the right

#combined....for the same function but in one line
df_call['Last_Name'] = df_call['Last_Name'].str.rstrip('_/.')
#.strip('_') instead of rstrip or lstrip will remove all instances of _
df_call
```

```python
df_call['Phone_Number'] = df_call['Phone_Number'].astype(str)
df_call
```

```python
#\D matches to any non-digit character and is replaced by nothing
df_call['Phone_Number'] = df_call['Phone_Number'].replace(to_replace=r'\D', value='', regex=True)
df_call
```

```python
df_call = df_call[df_call != 'N/a']
df_call = df_call.dropna()
df_call
```

```python
df_call['Paying Customer'] = df_call['Paying Customer'].replace(to_replace=r'^Y$', value='Yes', regex=True)
df_call
```

```python
df_call['Paying Customer'] = df_call['Paying Customer'].replace(to_replace=r'^N$', value='No', regex=True)
df_call
```

```python
df_call['Do_Not_Contact'] = df_call['Do_Not_Contact'].replace(to_replace=r'^Y$', value='Yes', regex=True)
df_call
```

```python
df_call['Do_Not_Contact'] = df_call['Do_Not_Contact'].replace(to_replace=r'^N$', value='No', regex=True)
df_call
```

```python
df_call = df_call.reset_index(drop=True) #Drop prevents an extra index column being added or is deleted
df_call
```

```python
df_call.to_excel('Customer Call List cleaned.xlsx', index=False)
```
